# Data Wrangling 

## Carregamento dos dados

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set(color_codes=True)

df = pd.read_csv('../googleplaystore.csv', encoding='UTF-8', header=0, sep=',')
# quando usar delimiter='\t'
# qual seria o encoding mais indicado? 'ISO-8859-1' ou 'UTF-8'
df.head()
```

## Remover dados Irelevantes

```python
df.columns
```

```python
df = df.drop(['Reviews','Genres','Last Updated','Current Ver','Android Ver'], axis=1)
df.head()
```

## Ajustar os tipos de dados

```python
df.dtypes
```

```python
#df['Size'] = df['Size'].astype('float')
#comando apresentou erro devido a coluna price conter a palavra 'M'
```

```python
df['Size'] = df['Size'].str.replace('M','')
#df['Size'] = df['Size'].astype('float')
#comando apresentou erro devido a coluna price conter a palavra 'Varies with device'
```

```python
df['Size'] = df['Size'].str.replace('Varies with device','')
#df['Size'] = df['Size'].astype('float')
#erro: 'could not convert string to float:' 
```

```python
#df["Price"].astype('float')
#comando apresentou erro devido a coluna price conter o caracter "$"
```

```python
df['Price'] = df['Price'].str.replace('$','')
#df['Price'].astype('float')
#comando apresentou erro devido a coluna price conter a palavra "Everyone"
```

```python
df['Price'] = df['Price'].str.replace('Everyone','')
#df['Price'].astype('float')
#erro: 'could not convert string to float:' 
```

## Renomear as Colunas

## Remover dados duplicados

```python
df.info()
```

```python
df = df.drop_duplicates()
df.head()
```

```python
df.info()
```

## Remover dados nulos

```python
print(df.isnull().sum())
df.count()
```

## Identificar e tratar os Outliers

```python
sns.boxplot(x=df['Rating'])
Q1=df['Rating'].quantile(0.25)
Q3=df['Rating'].quantile(0.75)
```

```python
Q1=df.quantile(0.25)
Q3=df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
df = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
df.shape
```

```python
sns.boxplot(x=df['Rating'])
Q1 = df['Rating'].quantile(0.25)
Q3 = df['Rating'].quantile(0.75)
print("Primeiro Quaril", Q1)
print("Terceiro Quaril", Q3)
```
